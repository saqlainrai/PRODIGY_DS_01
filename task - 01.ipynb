## Import Main Library

```python

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

## Read Dataset 

```python
df = pd.read_csv('countries-table.csv')
df.head(50)
```

## Data Visulization

```python
# Line plot of population over time
plt.plot(df['pop1980'], df['pop2050'])
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('World Population Over Time')
plt.show()
```

```python
# Bar plot of top 10 countries with highest population increase
top_countries_increase = df.nlargest(10, 'netChange')
plt.bar(top_countries_increase['country'], top_countries_increase['netChange'])
plt.xlabel('Country')
plt.ylabel(' Population Change')
plt.title('Top 10 Countries with Highest Population Increase')
plt.xticks(rotation=45)
plt.show()
```

```python
#Top 10 most populated countries in pop 1980
top_10_countries = df.nlargest(10, 'pop1980')
plt.figure(figsize=(12, 6))
sns.barplot(x='country', y='pop2023', data=top_10_countries)
plt.title('Top 10 Countries by Population (1980)')
plt.xlabel('Country')
plt.ylabel('Population (in billions)')
plt.xticks(rotation=45)
plt.show()
```

```python
countries = ['India', 'Egypt', 'United States', 'Iran', 'Japan', 'Iraq', 'Yemen', 'Argentina']

years = ['pop1980', 'pop2000', 'pop2010', 'pop2023','pop2030','pop2050']  

plt.figure(figsize=(10, 6)) 

for country in countries:
    country_data = df[df['country'] == country]
    plt.plot(years, country_data[years].values.reshape(-1), label=country)

plt.xlabel('Year')
plt.ylabel('Population')
plt.title('Population Trend for Selected Countries')
plt.legend()
plt.show()
```

```python
country_name = 'Iraq'
plt.bar(country_data.columns[14:16], country_data.iloc[:, 14:16].values.flatten())
plt.xlabel('Year')
plt.ylabel('Fertility Rate')
plt.title(f'Fertility Rate Comparison for {country_name}')
plt.show()
```

```python
# Population Growth Rate Distribution
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='growthRate', bins=20, kde=True)
plt.title('Population Growth Rate Distribution')
plt.xlabel('Growth Rate (%)')
plt.ylabel('Count')
plt.show()
```
